```python
# Upload the dataset 
dataset = pd.read_csv('testData.csv')
dataset.head() 
```

```python
# Process the dataset 
dataset = dataset.dropna()
dataset = dataset[['Open', 'High', 'Low', 'Close']]
dataset.head()
```

```python
# MA = Moving average is the average of stock's close price over a certain period of time (We are considering 3/ 10/ 30 days)
# RSI = Relative Strength Index/ Momentum feature/ Measures the speed and change of price movements
# Williams %R = Williams Percent Range 
# Williams %R = It compares a stock's closing price to the high-low range over a specific period, typically 14 days or periods
# Williams %R = an indicator that oscillates between 0 and -100, providing insight into the weakness or the strength of a stock. 

dataset['3day MA'] = dataset['Close'].shift(1).rolling(window = 3).mean()
dataset['10day MA'] = dataset['Close'].shift(1).rolling(window = 10).mean() 
dataset['30day MA'] = dataset['Close'].shift(1).rolling(window = 30).mean() 

dataset['Williams %R'] = talib.WILLR(dataset['High'].values, dataset['Low'].values, dataset['Close'].values, 14)
dataset['RSI'] = talib.RSI(dataset['Close'].values, timeperiod = 14) 
```

```python
dataset = dataset.dropna()
# Removes any rows with Null Values
```

```python
data = dataset.iloc[:, :] 
```

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Close'], color='r', label='Close Prices')
plt.plot(data['3day MA'], color='g', label='3-Day MA')
plt.title("Tata Motors STOCK - Close Vs 3-Day MA") 
plt.legend() 
plt.show() 
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Close'], color='r', label='Close Prices')
plt.plot(data['10day MA'], color='b', label='10-Day MA')
plt.title("Tata Motors STOCK - Close Vs 10-Day MA") 
plt.legend() 
plt.show() 
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Close'], color='r', label='Close Prices')
plt.plot(data['30day MA'], color='m', label='30-Day MA')
plt.title(" Tata Motors STOCK - Close Vs 30-Day MA") 
plt.legend() 
plt.show() 
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Close'], color='r', label='Close Prices')
plt.plot(data['RSI'], color='y', label='RSI')
plt.title("Tata Motors STOCK - Close Vs RSI") 
plt.legend() 
plt.show() 
```

```python
plt.figure(figsize=(15,5))
plt.plot(data['Close'], color='r', label='Close Prices')
plt.plot(data['Williams %R'], color='c', label='Williams %R') 
plt.title("Tata Motors STOCK - Close Vs Williams %R") 
plt.legend() 
plt.show() 
```
